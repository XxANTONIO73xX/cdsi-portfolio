# Procesamiento de imagenes

```python
import cv2
import matplotlib.pyplot as plt

imagen = cv2.imread('imagen.webp')

# Convertir la imagen de BGR a RGB (Matplotlib espera el formato RGB)
imagen_rgb = cv2.cvtColor(imagen, cv2.COLOR_BGR2RGB)

# Mostrar la imagen utilizando Matplotlib
plt.imshow(imagen_rgb)
plt.title('Imagen Original')
plt.show()
print(imagen.shape)
```

### Redimensionamiento de una imagen

```python
nueva_anchura = 200
nueva_altura = 200
imagen_redimensionada = cv2.resize(imagen_rgb, (nueva_anchura, nueva_altura))

plt.imshow(imagen_redimensionada)
plt.title('Imagen Redimensionada')
plt.show()
print(imagen_redimensionada.shape)
```

### Rotacin de una imagen

```python
angulo = 45

matriz_rotacion = cv2.getRotationMatrix2D((imagen_redimensionada.shape[1] / 2, imagen_redimensionada.shape[0] / 2), angulo, 1.0)

imagen_rotada = cv2.warpAffine(imagen_redimensionada, matriz_rotacion, (imagen_redimensionada.shape[1], imagen_redimensionada.shape[0]))

plt.imshow(imagen_rotada)
plt.title('Imagen Rotada')
plt.show()
```

### Conversin a escala de grises


```python
imagen_gris = cv2.cvtColor(imagen_rgb, cv2.COLOR_RGB2GRAY)

plt.imshow(imagen_gris, cmap='gray')
plt.title('Imagen en Escala de Grises')
plt.show()
```

### Umbralizacin de una imagen:

```python
umbral, imagen_umbralizada = cv2.threshold(imagen_gris, 128, 255, cv2.THRESH_BINARY)

plt.imshow(imagen_umbralizada, cmap='gray')
plt.title('Imagen Umbralizada')
plt.show()
```

### Deteccin de contornos

```python
contornos, _ = cv2.findContours(imagen_umbralizada, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

imagen_con_contornos = imagen.copy()
cv2.drawContours(imagen_con_contornos, contornos, -1, (0, 255, 0), 2)

imagen_con_contornos_rgb = cv2.cvtColor(imagen_con_contornos, cv2.COLOR_BGR2RGB)

plt.imshow(imagen_con_contornos_rgb)
plt.title('Imagen con Contornos')
plt.show()
```
